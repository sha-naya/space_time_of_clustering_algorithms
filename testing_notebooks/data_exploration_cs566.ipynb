### Obesity data
#### Medium size, real life data

```python
obesity_dataset = pd.read_csv(
    filepath_or_buffer='/Users/ayan/Downloads/ObesityDataSet_raw_and_data_sinthetic (2)/ObesityDataSet_raw_and_data_sinthetic.csv'
)
obesity_dataset.head()
```

```python
obesity_dataset.loc[obesity_dataset['NObeyesdad'] == 'Obesity_Type_I', 'NObeyesdad'] = 'Obesity'
obesity_dataset.loc[obesity_dataset['NObeyesdad'] == 'Obesity_Type_II', 'NObeyesdad'] = 'Obesity'
obesity_dataset.loc[obesity_dataset['NObeyesdad'] == 'Obesity_Type_III', 'NObeyesdad'] = 'Obesity'

obesity_dataset.loc[obesity_dataset['NObeyesdad'] == 'Overweight_Level_I', 'NObeyesdad'] = 'Overweight'
obesity_dataset.loc[obesity_dataset['NObeyesdad'] == 'Overweight_Level_II', 'NObeyesdad'] = 'Overweight'

obesity_dataset.NObeyesdad.value_counts()
```

```python
gender_dummies = pd.get_dummies(data=obesity_dataset.Gender)
family_history_dummies = pd.get_dummies(data=obesity_dataset.family_history_with_overweight, prefix='fam_his_of_obesity')
favc_dummies = pd.get_dummies(data=obesity_dataset.FAVC, prefix='favc')
caec_dummies = pd.get_dummies(data=obesity_dataset.CAEC, prefix='caec')
smoke_dummies = pd.get_dummies(data=obesity_dataset.SMOKE, prefix='smoke')
scc_dummies = pd.get_dummies(data=obesity_dataset.SCC, prefix='scc')
calc_dummies = pd.get_dummies(data=obesity_dataset.CALC, prefix='calc')
mtrans_dummies = pd.get_dummies(data=obesity_dataset.MTRANS, prefix='mtrans')

frames = [
    obesity_dataset,
    gender_dummies,
    family_history_dummies,
    favc_dummies,
    caec_dummies,
    smoke_dummies,
    scc_dummies,
    calc_dummies,
    mtrans_dummies,
]

obesity_dataset_with_dummies = pd.concat(objs=frames, axis=1)
obesity_dataset_with_dummies.head()
```

```python
columns_to_drop = [
    'Gender',
    'family_history_with_overweight',
    'FAVC',
    'CAEC',
    'SMOKE',
    'SCC',
    'CALC',
    'MTRANS',
]
formatted_obesity_dataset = obesity_dataset_with_dummies.drop(columns=columns_to_drop)
classification_column = formatted_obesity_dataset.pop('NObeyesdad')
formatted_obesity_dataset.insert(loc=0, column='NObeyesdad', value=classification_column)
formatted_obesity_dataset.head()
```
